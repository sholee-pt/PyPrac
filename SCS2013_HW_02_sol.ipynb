## HW 2-1. Function (25pts)

Instruction:

**(25pt)** 다음과 같은 함수 `check_lst`를 구현하세요:
* 입력: 2차원 리스트 `lsts`를 입력으로 가집니다
  * 2차원 리스트는 리스트를 element로 가지는 리스트입니다
  * 예시: `[[1,2,3],[4,5],[6,7,8,9]]`

* 기능: 
  * 함수의 입력값인 `lsts`의 type이 list인지 확인하여 list가 아닐 시 `Not a list!`를 출력하고 `False`를 return합니다.
  * 입력으로 받은 `lsts`가 list라면, 내부 element에 해당하는 리스트들이 같은 "길이"를 갖는지 확인하여 길이가 다르다면 `Lists inside have different size`를 출력하고 `False`를 return합니다.
    * `lsts` 내부의 원소들은 모두 리스트라고 가정합니다.

  * `lsts`의 type이 list이고 내부 element에 해당하는 리스트들의 길이가 모두 같다면 `True`를 return합니다. 

* 주어진 2차원 리스트 `lsts1`, `lsts2`, `lsts3`를 각각 함수의 입력으로 전달하여 동작을 확인하세요. 


**실행 결과** | Result:
```
>>>
---- CHECK1 ----
True
---- CHECK2 ----
Lists inside have different size
False
---- CHECK3 ----
Not a list!
False
```


````python
lst1 = [12, 13, 13, 23, 32, 33, 24, 22, 15, 11]
lst2 = [14, 18, 5, 40, 5, 38, 30, 14, 11, 13]
lst3 = [23, 8, 12, 12, 6, 32, 7, 9, 40, 39]
lst4 = [37, 14, 27, 36, 9, 13, 4, 20, 10, 20, 40, 60, 70, 13]

lsts1 = [lst1, lst2, lst3]
lsts2 = [lst1, lst2, lst3, lst4]
lsts3 = 3
````

````python
# 1. 
# sol:
def check_lst(lsts):
  if type(lsts) != list:
    print('Not a list!')
    return False

  lst_len = len(lsts[0])
  for lst in lsts:
    if len(lst) != lst_len:
      print('Lists inside have different size')
      return False

  return True
````

````python
# result:
print("---- CHECK1 ----")
print(check_lst(lsts1))
print("---- CHECK2 ----")
print(check_lst(lsts2))
print("---- CHECK3 ----")
print(check_lst(lsts3))
````

## HW 2-2. Function (25pts)

Instruction:

**(25pt)** 2차원 리스트를 입력으로 가지는 함수 `sum_elemwise_lst`를 구현하세요.

* `sum_elemwise_lst`는 2차원 리스트 `lsts`를 입력으로 가집니다. 
* `lsts`의 내부에 속한 리스트들을 **element-wise로 더하여 얻은 리스트를 return** 하는 함수입니다.
  * 예시: `sum_elemwise_lst([[1,5,4],[6,12,5],[1,2,3]]) --> [8, 19, 12]`

* 이를 위해서는 2-1에서 만든 `check_lst` 함수를 사용하여 `lsts`가 길이가 같은 리스트들로 이루어진 2차원 리스트인지 확인하여야 합니다: 
  * `check_lst`로 확인한 결과가 `False`인 경우 element-wise sum 기능을 제대로 수행할 수 없으므로 "-1"을 return합니다.

* 반복문을 사용하여 element-wise sum을 구현하여 결과 리스트를 return 하세요. 

**실행 결과** | Result:
```
>>>
---- SUM 1 ----
[49, 39, 30, 75, 43, 103, 61, 45, 66, 63]
---- SUM 2 ----
Lists inside have different size
-1
---- SUM 3 ----
Not a list!
-1
```

````python
# 2.
# sol:

def sum_elemwise_lst(lsts):
  if not check_lst(lsts):
    return -1
  
  result = []
  lst_len = len(lsts[0])

  for i in range(lst_len):
    sum = 0
    for j in range(len(lsts)):
      sum += lsts[j][i]
    result.append(sum)

  return result
````

````python
# result:
print("---- SUM 1 ----")
print(sum_elemwise_lst(lsts1))
print("---- SUM 2 ----")
print(sum_elemwise_lst(lsts2))
print("---- SUM 3 ----")
print(sum_elemwise_lst(lsts3))
````

## HW 2-3: Dice Game (50pt)
Instruction:

1. **(50pt)** 조건문, 반복문과 함수를 사용하여 아래의 게임을 구현하세요:

이 게임에서는 두 명의 플레이어가 번갈아가며 주사위를 던져 **목표 점수(goal_score)를 먼저 달성**하기 위한 경기를 진행합니다. 

이 게임을 플레이하는 `play_dice`라는 함수를 작성해야 합니다.
- 목표 점수 `goal_score`를 입력으로 가집니다.

- 주사위는 일반적인 '6면 주사위'로 `1`부터 `6`까지의 눈금을 가집니다.

- 홀수가 나오면 해당 눈금 만큼의 점수가 감소되는 penalty가 적용됩니다. 
  * 예를 들어 `1`이 나오면 플레이어의 현재 점수가 `1`만큼 감소하고, `3`이 나오면 `3`만큼 감소하며, `5`가 나오면 `5`만큼 감소합니다. 
  * 그러나 점수가 `0` 미만으로는 내려가지 않습니다.
- 짝수가 나오면 해당 눈금만큼 점수가 증가합니다.
- `goal_score`에 도달하면 승자를 출력하고 게임을 종료합니다.


**note**
- 사용자의 점수가 유효한지 확인해야 할 수도 있습니다: 0보다 크거나 같은지 확인해야 합니다.
- 사용자가 던진 숫자가 홀수인지 확인해야 할 수도 있습니다.
- 어느 한 플레이어의 점수가 목표 점수에 도달하면 게임을 멈추고 승자가 출력됩니다.
- 꼭 코드 셀의 구조를 따를 필요는 없습니다 (자유롭게 작성해도 됩니다)

````python
import random
# 꼭 아래의 구조를 따를 필요는 없음!

# sol:
def play_dice(goal_score):
  '''
  An interactive two-person game; also known as Dice game.
  @param goal_score: The target score the player needs to win a game.
  @function random.randint(1,6) can generate a random dice result every time it is called
  '''
  player1_score = 0
  player2_score = 0

  while True:
    # player 1
    #-------------------------------------
    input("Player 1's turn to roll a dice. Press Enter:") # 게임의 진행을 위해 엔터를 입력하도록 함
    dice_result = random.randint(1, 6)
    print("Player 1 rolled", dice_result)

    # check odd-number penalty for player 1
    if dice_result in [1,3,5]:
      player1_score -= dice_result
      player1_score = max(0, player1_score)
    else:
      player1_score += dice_result

    # check win or lose
    if player1_score >= goal_score:
      print('Player 1 wins!')
      break

    # check score of player 1
    print("Player 1's current score:", player1_score)
        
    # player 2
    #-------------------------------------
    input("Player 2's turn to roll a dice. Press Enter:")
    dice_result = random.randint(1, 6)
    print("Player 2 rolled:", dice_result)

    # check odd-number penalty for player 2
    if dice_result in [1,3,5]:
      player2_score -= dice_result
      player2_score = max(0, player2_score)
    else:
      player2_score += dice_result

    # check win or lose
    if player2_score >= goal_score:
      print('Player 2 wins!')
      break
    
    # check score of player 2
    print("Player 2's current score:", player2_score)

  print('End of a game!')
````

````python
# result:
play_dice(7)
````
