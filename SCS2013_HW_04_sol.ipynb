## HW 4-1. Regular Expressions (50 pts)
Instruction:

**1. (15pt)** 정규 표현식을 사용하여 다음 문제를 해결합니다.
아래에 제공된 리스트 `data`의 모든 숫자를 (정수, 소수 포함) 찾습니다. `data` 내부의 모든 요소는 $0$ 이상의 값을 가진다고 가정합니다.

**실행 결과** | Result:
```
>> 
['123.1', '4567', '789', '0.15', '257']
```

````python
# 1 
import re

data = ['apple', '123.1', 'orange', '4567', 'banana', '789', '0.15AB257']

# ========== your code here ==========
# sol: 
num_list = []

for val in data:
  num_list += re.findall(r'\d+\.?\d+', val)

# sol2:
num_list = re.findall(r'\d+(?:\.\d+)?', ' '.join(data))
# ==========
print(num_list)
````

**2. (20pt)** 정규 표현식을 사용하여 아래 요구 사항에 따라 비밀번호를 매칭하고 매칭 결과 `True/False`를 출력합니다.

  - 입력한 비밀번호 길이는 `6-12`자리 여야 하며(`6과 12 포함`), `영문자`와 `숫자만` 입력할 수 있습니다.

**실행 결과** | Result: ({}는 사용자의 입력을 의미)
```
>> Enter your password: {123456}
   True
>> Enter your password: {QwerTyuiopas}
   True
>> Enter your password: {123Wert}
   True
>> Enter your password: {123w}
   False
>> Enter your password: {@#$%$%no}
   False
```
  

````python
# 2
import re

password = input('Enter your password: ')

# ========== your code here ==========
# sol:
pattern = r'^[a-zA-Z0-9]{6,12}$'
mat = re.match(pattern, password)
print(bool(mat))
# ==========
````

**3. (15pt)** 아래에 주어진 리스트 `text`에서 각 string element마다 매 단어의 첫 글자를 따 약자를 생성하고자 합니다. 각 string element의 약자를 생성하여 `abbreviation` 리스트에 저장하여 출력하세요.

  - ex) World Health Organization -> WHO    

  
**실행 결과** | Result:
```
>>
['FEMA', 'IRA', 'DUP', 'FaDA', 'OoLC']
```
  

````python
# 3
import re

text = [
  'Federal Emergency Management Agency',
  'Irish Republican Army',
  'Democratic Unionist Party',
  'Food and Drug Administration',
  'Office of Legal Counsel'
]

# ========== your code here ==========
# sol:
abbreviations = []
pattern = r'\b(\w)'

for phrase in text:
  result = re.findall(pattern, phrase)
  abb_word = ''.join(result)
  abbreviations.append(abb_word)

# sol2:
for phrase in text:
    matches = re.finditer(pattern, phrase)
    abbreviation = ''.join(match.group() for match in matches)
    abbreviations.append(abbreviation)
# ==========
print(abbreviations)
````
