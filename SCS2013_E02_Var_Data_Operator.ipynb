### Arithmetic Operators 

Used for most basic mathematical operations 

* `+` Addition (더하기): adds values 
* `-` Subtraction (빼기): subtracts values
* `*` Multiplication (곱하기): multiplies values
* `/` Division (float) (나누기): divides the first value by the second value and returns value with decimals
* `//` Division (floor) (몫): divide two values and returns integers without decimals
* `%` Modulus (나머지): gives the remainder of a division result
* `**` Power (거듭제곱): raises the first value to the power of second value 

````python
a = 5
b = 3

c = a + b
d = a - b
e = a * b
f = a / b
g = a // b
h = a % b
i = a ** b

print('a is', a, 'and b is', b)
print('a + b is', c)
print('a - b is', d)
print('a * b is', e)
print('a / b is', f)
print('a // b is', g)
print('a % b is', h)
print('a ** b is', i)
````

````python
# division operator returns value with decimals (항상 계산 결과를 소수점이 있는 실수 형태로 반환)
print(10/1)
print(10/2)
print(10/3)
print(10/4)
print(10/5)
````

### Comparison Operators 

* returns "**boolean** values (True or False)" of the comparison results
* `==` Equal: returns True if left side is equal to the right side
* `!=` Not equal: returns True if left side is not equal to right side
* `>` Greater than: returns True if left side is grater than the right side
* `<` Less than: returns True if left side is less than the right side
* `>=` Greater than or equal to: returns True if left side is grater than or equal to the right side
* `<=` Less than: returns True if left side is less than or equal to the right side

````python
a = 5
b = 3

c = a == b
d = a != b
e = a > b
f = a < b
g = a >= b
h = a <= b

print('a is', a, 'and b is', b)
print('a == b is', c)
print('a != b is', d)
print('a > b is', e)
print('a < b is', f)
print('a >= b is', g)
print('a <= b is', h)
````

### Logical Operators 
* returns "**boolean** values (True or False)" 
* `and` logical AND: returns True when both operatnds are True (논리곱: 두개의 명제가 모두 참인지 판단)
* `or` logical OR: returns True if at least one of the operands is True (논리합: 적어도 한개 이상의 참이 있는지 판단)
* `not` logical NOT: returns True if operand is False

````python
x = 5
print(x > 3 and x < 10)
````

````python
x = 15
print(x > 3 and x < 10)
````

````python
x = 5
print(x > 3 or x < 4)
````

````python
x = 5
print(x < 2 or x > 6)
````

### Membership Operators / Identity Operators 

* returns "**boolean** values (True or False)" 
* `in`: return True if an object is in another object
* `not in`: return True if an object is not in another object

* `is`: tells if two objects are identical
* `is not`: tells if two objects are not identical

**Note** that `is` returns False if it is not the same object, even if they have the same content 


````python
x = ['apple', 'banana']
print('apple' in x)
print('pear' in x)
````

````python
a = 10
b = 3
A = {1,2,3,4,5}
print(a in A)
print(b in A)
````

````python
a, b = 5, 5
print(id(a), id(b))
print(a is b)
print(a == b)
````

````python
a, b = 56, 92
print(id(a), id(b))
print(a is b)
print(a == b)
````

````python
x = 1000
print(x, id(x))
y = 1001
print(y, id(y))
y = y-1
print(y, id(y))

print(x is y)
print(x == y)
````

### More Assignment Operators 

* `+=`: adding values and assigning it: `a+=b` instead of `a=a+b`
* Similarly, we can use `-=`, `*=`, `/=`, `%=`, `//=`, `**=`, ... 

````python
x = 5
print('x is', x)

x += 3
print('x += 3 is', x)

x -= 2
print('x += 2 is', x)

x *= 1.5
print('x *= 1.5 is', x)

x /= 1.25
print('x /= 1.25 is', x)

x //= 3
print('x //= 3 is', x)

x **= 4
print('x **= 4 is', x)
````
