## Create Strings, String Operators

* Python strings have powerful text-processing features. Strings are created by using **quotes (따옴표)** with texts in between. 
The quotes used can be either 
  - i) single **`' '`** 작은따옴표
  - ii) double **`" "`** 큰따옴표
  - iii) three single quotes **`'''  '''`** 작은따옴표 3개 연속
  - iv) three double quotes **`""" """`** 큰따옴표 3개 연속

* Escape character (이스케이프 코드): backslash `\` 뒤에 문자가 오는 조합
  - `\'`, `\"`: single quote, double quote
  - `\n`, `\t`: new line, tab
  - `\\`: backslash

## String Indexing, Slicing and Striding

* **Indexing**: `my_str[index]` 
  - starts from `0` - from left to right 
  - 0부터 시작하여 각 문자의 위치를 표현
  - negative index starts from `-1` - from right to left

* **Slicing**: `my_str[start index: end index]` 
  - it includes the character at `start index`, but not the character at `end index`
  - `start index`, `end index`를 생략할 수 있음

* **Striding**: `my_str[start index: end index: striding number]` 
  - it also includes the character at `start index` but not the character at `end index`
  - `striding number`만큼 건너뛰면서 선택

````python
my_str = "Hello World"
print(my_str)
````

````python
# my_str[index]
print(my_str[0])
print(my_str[4])
print(my_str[9])

print(my_str[-1])
print(my_str[-7])
````

````python
# my_str[start index : end index]
print(my_str)

print(my_str[0:3])
print(my_str[4:-2])
print(my_str[1:4] + my_str[-5:9])
````

````python
# my_str[start index : end index]
print(my_str[:7])
print(my_str[5:])
````

````python
# my_str[start index : end index: stride]
print(my_str[0:7:2])
print(my_str[::3])
````

## String Formatting: **% operator**, **.format()**, **f-string** methods

Let's check out some exercises to uderstand `% operator`, `.format()`, `f-string` methods better.

* % operator: format specifier를 사용하여 특정 부분을 원하는 형식으로 표현 
  - 문자열 내에 원하는 위치에 포맷 코드를 사용
  - `%s`: string, `%c`: character, `%d`: integer, `%f`: floating point, `%%`: % character
  - 문자열을 다 만든 후 (따옴표 이후에) `%`를 통해 삽입하고자 하는 값 혹은 변수를 명시 

* .format(): `{}`와 `.format()`을 사용해 원하는 형식을 표현
  - 문자열 내에 `{}`으로 원하는 위치를 명시
  - 문자열을 다 만든 후 `.format()` 함수를 사용해 삽입하고자 하는 값 혹은 변수를 전달 
  - `{}`에 0으로 시작하는 index 순서를 매기거나, 이름을 붙일 수 있음 

* f-string: `f' '`와 `{}`을 사용해 원하는 형식을 표현
  - 문자열 앞에 `f`를 붙여 `f-string`을 생성
  -문자열 내에 `{}`으로 원하는 위치를 명시하고, `{}` 내에 삽입하고자 하는 값 혹은 변수를 직접 참조 
  
* 정렬, 공백 채우기, 자리수 지정하기
  - `:<10`, `:>10`, `:^10`: 왼쪽, 오른쪽, 가운데 정렬
  - `:a<10`, `:a>10`, `:a^10`: 정렬 & 공백을 a로 채움
  - `:A.Bf`: 전체 자릿수 A (소수점 . 을 포함), 소수점 아래 B자리

### Multiple values

If we want to fill multiple values, we can use multiple curly brackets or multiple % operators.

Print the numbers `1`, `3.14` and string `Python`, each separated with a comma.

````python
my_str = 'Print the values: %d, %f and %s' %(1, 3.14, 'Python')
print(my_str)
````

````python
my_str = 'Print the values: {0}, {1} and {2}'.format(1, 3.14, 'Python')
print(my_str)
````

````python
my_str = 'Print the values: {v1}, {v2} and {v3}'.format(v2=3.14, v1=1, v3='Python')
print(my_str)

vala = 1
valb = 3.14
valc = 'Python'

my_str = 'Print the values: {v1}, {v2} and {v3}'.format(v2=valb, v3=valc, v1=vala)
print(my_str)
````

````python
my_str = f'Print the values: {vala}, {valb} and {valc}'
print(my_str)
````

### How to format numbers

A common problem with printing numbers is specifying the number of decimal places. Try these examples:

````python
no = 23.53206945
print(no)
````

````python
print('!%.4f!' % no)
print('!%15.2f!' % no)
````

````python
print(f'1: {no}')
print(f'2: {no:f}')
print(f'3: {no:0.4f}')
print(f'4: {no:>15.2f}')
print(f'5: {no:<15.2f}')
print(f'6: {no:*>15.2f}')
````

````python
day = 'Today'
print('{}\'s temperature: {:d}°C'.format(day, 100))
print('{1}\'s temperature: {0:d}°C'.format(100, day))
print('{a}\'s temperature: {b:.3f}°C'.format(b=100, a=day))

print('{}\'s temperature: {:>10.3f}°C'.format(day, 100))
print('{aa}\'s temperature: {bb:<10.3f}°C'.format(aa=day, bb=100))
print('{}\'s temperature: {:^10.3f}°C'.format(day, 100))
````

## String Methods

There are many in-built string methods in Python. Let's try out some exercises!
- example: `my_str.count('a')`

**Note**: All string methods returns new values. They do not change the original string!!
- 문자열 내장함수는 새로운 문자열을 만들어 반환함: 원래 문자열의 값이 바로 바뀌지 않음 

````python
my_str = "Hello World. Attention please."
print(my_str)
````

````python
# .count()
print(my_str.count('l'))
print(my_str.count('e'))
print(my_str.count('ea'))
print(my_str)
````

````python
# .find(), .index()

print(my_str.find('o'))
print(my_str.find('lo'))
print(my_str.find('y'))

print(my_str.index('o'))
print(my_str.index('lo'))
#print(my_str.index('y'))
````

````python
# .replace()

print(my_str.replace('World','Guys'))
print(my_str)

new_str = my_str.replace('Hello', 'Hi')
print(my_str)
print(new_str)
````

````python
# .split()

print(my_str)
print(my_str.split())
print(my_str.split('l'))
````

````python
# .split()
str2 = 'Ally, Bob, Charlie, David, John'
str2.split(',')
````

````python
# .strip(), .lstrip(), .rstrip()

str3 = '   Hello   hi!   '
print(str3)

print(str3.strip())
print(str3.lstrip())
print(str3.rstrip())
````

````python
# .upper(), .lower()

print(my_str.upper())
print(my_str.lower())
````

````python
# .join()

new_str = '+'.join(my_str)
print(my_str)
print(new_str)

str1 = 'aaa'
str2 = 'test'
print(str1.join(str2))
print(str2.join(str1))
````
