- `is_right_triangle()` method

  - **note**: `combinations()` 함수를 사용할 수 있습니다. 아래의 예시를 참고하세요. 혹은 loop를 사용해도 됩니다. 
  - `triangles`: 가능한 세 변의 길이의 조합 중 삼각형을 이루는 경우를 모두 골라 instance variable인 `triangles` 리스트를 완성합니다. 
  - `right_triangles`: 가능한 세 변의 길이의 조합 중 직각삼각형을 이루는 경우를 모두 골라 instance variable인 `right_triangles` 리스트를 완성합니다. 
  - 이 함수의 반환(return) 값은 가능한 삼각형의 총 개수와 직각삼각형의 총 개수입니다.

````python
# combinations() example
from itertools import combinations

items_1 = [1, 2, 3, 4]
# 길이가 3인 모든 조합을 찾아냅니다.
combs = combinations(items_1, 3)
for i in combs:
    print(i)

items_2 = ['a', 'b', 'c', 'b']
# 길이가 2인 모든 조합을 찾아냅니다.
combs = combinations(items_2, 2)
for i in combs:
    print(i)

# 그러나 이 함수는 모든 조합을 찾아내기 때문에 중복되는 조합이 발생할 수 있다는 점에 유의합니다.
````

````python
# sol:
from itertools import combinations

class TriangleCounter:
  def __init__(self, sides):
    self.sides = sides

  def is_triangle(self, tri_sides):
    a,b,c = sorted(tri_sides)
    return a + b > c
    # return a+b > c and b+c > a and c+a > b

  def is_right_triangle(self, tri_sides):
    a,b,c = sorted(tri_sides)
    return c**2 == a**2 + b**2
  
  def count_triangles(self):
    side_combs = combinations(self.sides, 3)
    self.triangles = []
    self.right_triangles = []

    for comb in side_combs:
      if self.is_triangle(comb) and (comb not in self.triangles):
        self.triangles.append(comb)
      if self.is_right_triangle(comb) and (comb not in self.right_triangles):
        self.right_triangles.append(comb)

    return len(self.triangles), len(self.right_triangles)

````

````python
# Test your code: example
sides = [15, 8, 14, 2, 10, 6]

counter = TriangleCounter(sides)
triangle_count, right_triangle_count = counter.count_triangles()

print(f"The given list can form {triangle_count} triangles, and {right_triangle_count} right triangles.")
print(counter.triangles)
print(counter.right_triangles)
````

````python
# sol:
class UserInfo:
  def __init__(self, minlen_id, maxlen_id, minlen_pw, maxlen_pw):
    self.minlen_id = minlen_id
    self.maxlen_id = maxlen_id
    self.minlen_pw = minlen_pw
    self.maxlen_pw = maxlen_pw
    self.infobank = {}

  def new_user(self, id, pw):
    flag = self.checkid(id) and self.checkpw(pw)

    if flag:
      self.infobank[id] = pw
      print('Success')
    else:
      print('Fail')

  def checkid(self, id):
    flag = self.minlen_id < len(id) < self.maxlen_id
    flag = flag and id not in self.infobank
    print(f'ID validness: {flag}')
    return flag

  def checkpw(self, pw):
    flag = self.minlen_pw < len(pw) < self.maxlen_pw
    print(f'PW validness: {flag}')
    return flag
````

````python
special_char = ['!', '@', '#', '$', '%', '^', '&', '*']

# sol:
class UserInfo_A(UserInfo):
  def __init__(self, minlen_id, maxlen_id, minlen_pw, maxlen_pw):
    super().__init__(minlen_id, maxlen_id, minlen_pw, maxlen_pw)
    self.special_char = ['!', '@', '#', '$', '%', '^', '&', '*']

  def checkpw(self, pw):
    # flag = super().checkpw(pw) and (len([char in pw for char in self.special_char]) >= 1)
    flag = (self.minlen_pw < len(pw) < self.maxlen_pw) and (sum([char in pw for char in self.special_char]) >= 1)
    print(f'PW validness: {flag}')
    return flag
````

**`UserInfo`를 부모로 하는 자식클래스 `UserInfo_B`를 구현하세요 (20pts)**
- 부모 클래스와 비슷한 동작을 하지만, PW를 암호화하여 저장하는 기능이 추가된 클래스입니다.
- `인스턴스 변수`
    - 부모 클래스로부터 상속
    - `encrpyt_key`: 클래스가 생성자가 받는 int. PW를 암호화하는데 사용됩니다.

- `new_user()` 메소드
    - `id` ; `pw`: 메소드의 매개변수, string 변수.  
    - `checkid()` 와 `checkpw()`를 통해 조건 `boolean`을 `return`받아, 조건이 만족(True)한다면 `info_bank` 딕셔너리에 `id`, `pw`를 key, value형태로 추가: 부모 클래스와 동일
    - 해당 메소드에서는 추가적으로, `encrypt()` 메소드를 호출하여 `pw`를 암호화합니다.

- `encrypt()` 메소드
    - `pw`: 메소드의 매개변수, string 변수.  
    - string의 각 character은 **ascii code**를 통해 숫자로 포현될 수 있습니다. ex) 'a' -> 97 ; 'b' -> 98 ; '!' -> 33
    - `ord(x)` 함수는 character `x`를 숫자로 변환해주는 함수며 `chr(y)` 함수는 숫자 `y`를 character로 변환해줍니다.
    - ex) `ord('a') -> 97 ; 'chr(97) -> 'a'`
    - `ord()` 함수와 `chr()` 함수를 이용하여 `pw`의 각 character를 숫자로 변환한 후, `encrypt_key` 정수만큼 더하여 이를 다시 character로 변환해 만든 새로운 문자로 `pw`를 암호화하세요.

- `checkid()` 메소드
    - 부모 클래스로부터 상속
    
- `checkpw()` 메소드
    - 부모 클래스로부터 상속

````python
# sol:
class UserInfo_B(UserInfo):
  def __init__(self, minlen_id, maxlen_id, minlen_pw, maxlen_pw, encrypt_key):
    super().__init__(minlen_id, maxlen_id, minlen_pw, maxlen_pw)
    self.encrypt_key = encrypt_key
  
  def new_user(self, id, pw):
    flag = self.checkid(id) and self.checkpw(pw)

    if flag:
      encrypted_pw = self.encrypt(pw)
      self.infobank[id] = encrypted_pw
      print('Success')
    else:
      print('Fail')
    
  def encrypt(self, pw):
    new_pw = ''
    for char in pw:
      new_pw += chr((ord(char) + self.encrypt_key))
    return new_pw
````
