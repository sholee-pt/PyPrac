## HW 1-3. List, If, Loop Statements (30pts)

Instruction:

1. **(15pt)** List의 정렬 함수 `.sort()`를 사용하지 않고, **반복문을 사용하여** `lst1`을 내림차순으로 정렬하는 코드를 작성하세요. 
  * 내림차순(descending order) 정렬: 큰 수부터 작은 수로 배치하는 것.
  * 반복문을 사용합니다.

2. **(15pt)** 반복문과 조건문을 사용하여, 주어진 숫자 `val`의 약수를 전부 구해 리스트로 만들어 출력하세요.
  * 약수(divisor): 나누어 떨어지게 하는 0이 아닌 정수

**실행 결과** | Result:
```
# 1
>>> 
A list sorted in descending order:  [100, 97, 90, 87, 81, 78, 72, 64, 50, 49, 29, 29, 4]

# 2 
>>> 
Divisor of 60: [1, 2, 3, 4, 5, 6, 10, 12, 15, 20, 30, 60]
```

````python
#1.
lst1 = [4, 49, 97, 29, 90, 72, 64, 29, 87, 50, 100, 81, 78]

# sol:
for i in range(len(lst1)-1):
  for j in range(i+1, len(lst1)):
    if lst1[i] < lst1[j]:
      temp_val = lst1[i]
      lst1[i] = lst1[j]
      lst1[j] = temp_val

print('A list sorted in descending order: ', lst1)
````

````python
#2.
val = 60

# sol:
divisor = []
for i in range(1, val+1):
  if val % i == 0:
    divisor.append(i)

print(f'Divisor of {val}: {divisor}')
````

## HW 1-4. Dictionary (25pts)

Instruction:

**(25pt)** `1000₩, 10000₩, 2$, 50$, 5€, 100€` 지폐로 구성된 지폐 더미에서 각 화폐의 지폐가 몇개씩 있는지 파악하고자 합니다. **딕셔너리와 반복문, 조건문**을 사용해 문제를 해결하세요.  

* 주어진 리스트 `money_lst`는 지폐 더미를 나타냅니다.
  * 각 지폐는 `2$` 와 같은 string 형태로 주어져 `money_lst` 리스트에 저장되어있습니다.
    * 예시: `money_lst: ['100€', '5€', '2$', '10000₩', '50$', '2$', '100€']`
  * 총 30개의 지폐를 random하게 생성하여 지폐 더미를 만들도록 되어 있습니다. 

* 반복문을 사용하여 각 지폐의 화폐 종류와 금액을 파악하고 갯수를 세어 정보를 딕셔너리 구조로 만듭니다.
  * 각 지폐권의 금액과 갯수를 세고자 하므로, 파이썬의 딕셔너리 구조를 이용합니다.
  * 화폐의 종류인 `$, ₩, €` 마다 각각 딕셔너리를 만듭니다: `bill_dollar`는 `$`에 해당하는 지폐의 정보를 저장하는 딕셔너리이며, `bill_won`은 `₩`에 해당하는 딕셔너리, `bill_euro`는 `€`에 해당하는 딕셔너리입니다. 
  * 각 화폐 종류마다 금액권이 다르므로, 각 금액의 지폐가 몇 장이 포함되어있는지 개수를 파악하여 딕셔너리로 만듭니다. 
    * 예시: `bill_dollar = {'2$':2, '50$':1}`

  * 조건문을 사용하여 `money_lst`의 element가 어떤 화폐를 나타내는지 확인하세요. (**hint:** `in, not in` 등을 사용할 수 있습니다)
  * 각 화폐에 해당하는 딕셔너리에 해당 금액권의 값을 key로 활용해 개수 정보를 업데이트합니다.
    * **note:** 처음 등장하는 금액권을 구별하여 처리할 필요가 있으니 주의하세요


* 지폐 더미의 모든 지폐 확인을 마쳤을 경우, 각 딕셔너리의 결과를 출력하여 지폐 개수를 확인하세요.
  * 난수 생성을 고정시켜두었으므로 아래와 동일한 결과가 나와야 합니다.

**실행 결과** | Result 
```
>>> 
Total dollar: {'2$': 5, '50$': 5}
Total won: {'10000₩': 5, '1000₩': 1}
Total euro: {'100€': 8, '5€': 6}
```

    

````python
import random # 난수를 생성하기 위한 모듈입니다.
random.seed(0) # 매 실행마다 같은 난수가 생성되도록 합니다.

# ₩₩로 바꿔서 하는 것이 안전합니다.
money_lst = random.choices(('1000₩₩', '10000₩₩', '2$', '50$', '5€', '100€'), k=30) # 25개의 지폐를 랜덤으로 생성하여 지폐 더미를 만듭니다.
print(f'Pile of {len(money_lst)} bills: {money_lst}')


# sol:
bill_won = {}
bill_dollar = {}
bill_euro = {}

for i in range(len(money_lst)):
  if '$' in money_lst[i]:
    if money_lst[i] not in bill_dollar.keys():
      bill_dollar[money_lst[i]] = 1
    else: 
      bill_dollar[money_lst[i]] += 1
  elif '₩₩' in money_lst[i]:
    if money_lst[i] not in bill_won.keys():
      bill_won[money_lst[i]] = 1
    else:
      bill_won[money_lst[i]] += 1
  elif '€' in money_lst[i]:
    if money_lst[i] not in bill_euro.keys():
      bill_euro[money_lst[i]] = 1
    else:
      bill_euro[money_lst[i]] += 1
  

print(f'Total dollar: {bill_dollar}')
print(f'Total won: {bill_won}')
print(f'Total euro: {bill_euro}')
````
