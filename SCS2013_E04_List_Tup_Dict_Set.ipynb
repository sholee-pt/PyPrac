## Lists (리스트)

List is a collection which is **ordered (index 사용)** and **changeable**. It **allows duplicate** items
  - it is a data sequence that can hold multiple, different types of data 

  - represented inside **brackets "[]"**
  - elements are separated with **commas ","**
  - **Syntax**
  ```
  list_name = [element1, element2, ...]
  ```
  - mutability = changeable: the values inside a list are changeable
  - it has an index order that starts with `0` - we can access each element inside a list with its index number: `list_name[index]`
  







````python
# Create a list
lst = ['Hello World', 3, 2.5, ['Python', 'Programming'], True]
lst
````

````python
# List indexing, slicing and striding
# check the length of the list
print(len(lst))

# check the index and slicing 
print(lst[0])
print(lst[2])
print(lst[1:3])
````

````python
print(lst[-2])

# check the type of lst[-2]
print(type(lst[-2]))

# access the element of the nested list
print(lst[-2][0])
print(lst[-2][1])
# print(lst[-2][0][-1])
````

````python
# Mutability 
print(lst)

# change index 2 value: using '=' 
lst[2] = 1.4
lst[-2][1] = 'Welcome!'
print(lst)
````

**List Methods**

Here are some list methods:
- `.append(a)`: adding element `a` to a list
- `.insert(idx, a)`: inserting element `a` to a particular position `idx` of the list
- `.index(a)`: show the index of an element `a`
- `.clear()`: clear all elements of the list
- `.remove(a)`: removing an element `a` of the list
- `.sort()`: sorting the elements of the list
- `.reverse()`: reversing the order of elements
- `.count(a)`: counting how many elements `a` there are in the list
- `sum()`: sum all the elements in the list
- `min()`: show the element with minimum value in the list
- `max()`: show the element with maximum value in the list
- `del list[idx]`: delete an element at a particular position `idx` of the list

**note**: `.append()`, `.insert()`, `.clear()`, `.remove()`, `.reverse()`: list의 값을 바로 변경함 (mutability)

````python
lst_1 = ['Hello', 'World', 31, 2.7, True, '31']
print('Original list:', lst_1)

# .append()
lst_1.append("Append")
print('After .append():', lst_1)

# .insert()
lst_1.insert(3, "Insert")
print('After .insert():', lst_1)
````

````python
# .index()
lst_1 = ['Hello', 'World', 31, 2.7, True, 'True', '31']
print(lst_1)
print(lst_1.index(31))
````

````python
lst_1 = ['Hello', 'World', 31, 2.7, True, '31']
print('Original list:', lst_1)

# .clear()
lst_1.clear()
print('After .clear():', lst_1)

lst_1 = ['Hello', 'World', 31, 2.7, True, '31']
print('Original list:', lst_1)

# .remove()
lst_1.remove('World')
print('After .remove():', lst_1)
# what if removing again
# lst_1.remove('World')

# .pop()
lst_1.pop(3)
print('After .pop():', lst_1)

# del 
del lst_1[1]
print('After del:', lst_1)
````

````python
lst_1 = [4, 2, 6, 5, 2, 0, 4]
print('Original list:', lst_1)

# .reverse()
lst_1.reverse()
print('After .reverse():', lst_1)

# .sort()
lst_1.sort()
print('After .sort():', lst_1)
````

````python
lst_1 = [4, 2, 6, 5, 2, 0, 4]
print('Original list:', lst_1)

# .index()
print(lst_1.index(2))

# .count()
print(lst_1.count(2))

print(lst_1) # list is not changed
````

````python
# sum(), min(), max()
lst_1 = [3, 75, 6, 23, 3, 13]
print(lst_1)

print(sum(lst_1))
print(min(lst_1))
print(max(lst_1))

print(lst_1) # list is not changed
````

## Tuples (튜플)

Tuple is a collection which is **ordered** and **unchangeable**. It **allows duplicate** items
  - it is a data sequence that can hold multiple, different types of data 

  - represented inside **parenthesis "()"**
  - elements are separated with **commas ","**
  - **Syntax**
  ```
  tuple_name = (element1, element2, ...)
  ```

  - immutability = unchangeable: the values inside a tuple cannot be changed
  - it has an index order that starts with `0` - we can access each element inside a tuple with its index number



````python
# create a tuple
tup = ('Hello World', 3, 2.5, ['Python', 'Programming'], True) 
tup
````

````python
# 괄호를 생략해도 무방
tup = 3, 2.4, 5

print(type(tup))
print(tup)
````

````python
# tuple indexing, slicing and striding
tup = ('Hello World', 3, 2.5, ['Python', 'Programming'], True) 
print(tup)

# length of tuple
print(len(tup))

# indexing, slicing, ...
print(tup[0])
print(tup[3:])

print(tup[-2])
print(tup[-1])

print(tup[-2][0])
````

````python
# Immutability 
tup = ('Hello World', 3, 2.5, ['Python', 'Programming'], True) 
print(tup)

tup[2] = 1.4
print(tup)

del tup[1]
print(tup)
````

**Tuple Methods**

Here are some tuple methods:
- `.index()`: show the index of an element
- `.count()`: counting how many elements there are in the tuple
- `sum()`: sum all the elements in the tuple
- `min()`: show the element with minimum value in the tuple
- `max()`: show the element with maximum value in the tuple

**note**: 'tuple' object has no attribute `.append(), .insert(), .clear(), .remove(), .reverse()` (because of immutability)


````python
tup_1 = ('Hello', 'World', 31, 2.7, True, 'True', '31')
print('Original tuple: ', tup_1)

tup_1.reverse()
tup_1
````

````python
tup_1 = ('Hello', 'World', 31, 2.7, True, 'True', '31')
print('Original tuple: ', tup_1)

# .index()
print(tup_1.index(31))
print(tup_1.index('31'))
# tup_1.index(3.2)

# .count()
print(tup_1.count('True'))
````

````python
# sum(), min(), max()
tup_1 = (3, 75, 6, 23, 3, 13)
print('Original tuple:', tup_1)

print(sum(tup_1))
print(min(tup_1))
print(max(tup_1))
````

## Dictionaries (딕셔너리)

Dictionary is a collection which is **unindexed** (ordering is not significant!!) and **changeable**. It has **no duplicate** items.
  - it is a data sequence that can hold multiple **"key-value" pairs**

  - represented inside **curly brackets "{}"**
  - key-value pairs are separated with **commas ","**
  - key and value is spearated by a **colon ":"**
  - **Syntax**
  ```
  dictionary_name = {key1:value1, key2:value2, ...}
  ```
  - can **access each value with its corresponding key** 
  ```
  dict[key]
  dict.get(key)
  ```

  - immutability: the *key* inside a dictionary cannot be changed and also should be unique    
  - mutability: the *value* inside a dictionary is changeable 

````python
# create dictionary 
dct = {'name':'Kim', 'birth':['March',9], 'age':27}
print(dct)

# what if there are duplicated keys?
dct = {'name':'Kim', 'birth':['March',9], 'age':27, 'name': 'James'} 
print(dct)

# what if there are duplicated values?
dct = {'name':'Kim', 'birth':['March',9], 'age':27, 'nickname': 'Kim'}
print(dct)
````

````python
dct = {'name':'Kim', 'birth':['March',9], 'age':27}
print(dct)

# access values using their keys or .get()
print(dct['name'])
print(dct.get('name'))

print(dct['age'])
print(dct['birth'])

print(dct['birth'][0])
print(dct['birth'][1])
````

````python
# Add/Remove a key-value pair
dct = {'name':'Kim', 'birth':['March',9], 'age':27}
print('Original dict:', dct)

# Add a pair ('year', 1995)
dct['year'] = 1995
print('After adding a pair:', dct)

# Delete a pair ('age', 27)
del dct['age']
print('After deleting a pair:', dct)

# Change a value of a specific key-value pair 
dct['name'] = 'James'
print('After changing a pair:', dct)
````

**Dicionary Methods**

Here are some dictionary methods:
- `.keys()`: show the keys in a dictionary
- `.values()`: show the values in a dictionary
- `.items()`: create a (key, value) tuple 

- `.clear()`: clear the entire dictionary
- `.copy()`: copy a dictionary
- `.update()`: insert the specific items (key-value, or dictionary) to the dictionary



````python
# Example: check a list of dictionary's keys, values, and key-value pairs
dct = {'name':'Kim', 'birth':['March',9], 'age':27}
print('Original dict:', dct)

# .keys()
print(dct.keys())
print(type(dct.keys()))

dct_key_list = list(dct.keys())
print(dct_key_list)
````

````python
# .values()
print(dct.values())
print(type(dct.values()))

# .items()
print(dct.items())
print(type(dct.items()))
````

````python
# in, not in
print('name' in dct.keys())
print('name' in dct) 
print('Name' in dct.keys())

print('Kim' in dct)
print('Kim' in dct.values())
````

````python
dct = {'name':'Kim', 'birth':['March',9], 'age':27}
print('Original dict:', dct)

# .clear()
dct = {'name':'Kim', 'birth':['March',9], 'age':27}
dct.clear()
print('After .clear():', dct)

dct = {'name':'Kim', 'birth':['March',9], 'age':27}
print('Original dict:', dct)

# .copy()
c_dct = dct.copy()
print('Copied dict:', c_dct)

# update()
dct_2 = {'year': 1995, 'Language': 'Python'}
dct.update(dct_2)

print('After .update():', dct)
print('Copied dict:', c_dct)
````

## Sets (집합)

Set is a collection which is **unordered**, **unchangeable**, and **unindexed**. It has **no duplicate** items
  - represented using **curly brackets "{}"**
  - elements are separated with **commas ","**
  - **Syntax**
  ```
  set_name = {element1, element2, ... }
  ```

  - immutability = unchangeable: the values inside a set cannot be changed, but we can add/remove items inside a set
  - it does not have an index order - we can use `in` or `not in` to check an element is included in a set or not





````python
# create a set
set_1 = {'A', 2, 'B', 'C', 2, 3, 'C'} # duplicate
print(set_1)

# create an empty set: set()
set_2 = {} 
print(set_2)
print(type(set_2))

set_2 = set()
print(set_2)
print(type(set_2))
````

````python
# accesing items in a set
set_1 = {'A', 2, 'B', 'C', 2, 3, 'C'} 
print(set_1)

print(3 in set_1)
print('a' in set_1)
print('A' in set_1)
````

````python
# Add/Remove items
set_1 = {'A', 2, 'B', 'C', 2, 3, 'C'}
print('Original set:', set_1)

set_1.add(6)
print('After .add():', set_1)

set_1.update(['a','A'])
print('After .update():', set_1)

set_1.remove(3)
print('After .remove():', set_1)
````

**Set Methods**

Here are some set methods:
- `A.union(B)` or `A | B`: 합집합
- `A.intersection(B)` or `A & B`: 교집합
- `A.difference(B)` or `A - B`: 차집합
- `A.issuperset(B)` or `A >= B`: 포함관계




````python
# set methods
set_a = {1,2,3,5,7,8}
set_b = {1,2,4,6,7}
set_c = {1,2,3,5}

print(set_a.union(set_b))
print(set_a.intersection(set_b))
print(set_a - set_b)
print(set_a.issuperset(set_b))
print(set_a >= set_c)
````
