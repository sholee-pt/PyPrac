## If Statement: Conditional statements (조건문)

With conditional statements, we can control the flow of execution by using certain statements. **"if" statements** are very useful whenever we need a conditional execution.

**Syntax**
```
if <condition>:
  <statement>
```
- starts with **"if"**, followed by a ``<condition>``, and ends with colon **colon :**
- the ``<condition>`` is evaluated as a *Boolean*: either True or False
- the ``<statement>`` is a code block to be executed
- each line of ``<statement>`` must **indented**: 4 spaces or a tab
- `<condition>`이 True인 경우에만 들여쓰기 된 `<statement>`가 실행됨



````python
# if: 독립적인 조건문 3개 
x = 3
# x = 30
# x = -10

if x < 10:
  print('x is less than 10')

if x > 10:
  print('x is greater than 10')

if x > 2 and x < 5:
  print('x is in range (2,5)')
````

````python
# indentation
x = 1
y = 8

if x > y:
  print('x is %d and y is %d' %(x,y))  
print('x is greater than y')
````

### If-Elif-Else Statement

* **if-else**: Make more detailed conditional statements if we use **"else"** statement. 

```
if <condition>:
  <if statements>
else:
  <else statements>
```

- `<condition>`이 True인 경우 `<if statements>`만 실행 후 다음 코드 (들여쓰기가 해제된 코드) 로 넘어감
- `<condition>`이 False인 경우 `<else statements>`만 실행 후 다음 코드 (들여쓰기가 해제된 고드) 로 넘어감 
- `else` is optional! 
- `else` does not require a condition!


````python
# if-else

x = 1
# x = 20
threshold = 10

if x < threshold:
  print('If statement')
  print(f"x is less than {threshold}")
else:
  print('Else statement')
  print(f"x is greater than or equal to {threshold}")
````

````python
# if-else

isStudent = True
age = 25

if isStudent and (20 <= age <= 29): 
  print('I am a student in twenties')
else:
  print('I am NOT a student in twenties')
````

* **if-elif-else**: Can use an **"elif"** between "if" and "else" as an additional conditional statement. We can evaluate several conditional cases by using "elif". 

```
if <condition1>:
  <statement1>
elif <condition2>:
  <statement2>
elif <condition3>:
  <statement3>
else:
  <statement4>
```

- if ``<condition1>`` is True, the program quits after executing ``<statement1>``.
- if ``<condition1>`` is False and ``<condition2>`` is True, the program quits after executing ``<statement2>``.
- if ``<condition1>, <condition2>`` are False and ``<condition3>`` is True, the program quits after executing ``<statement3>``.
- otherwise (= all ``<condition1>, <condition2>, <condition3>`` are False), the program quits after executing  ``<statement4>``.
- `elif` is optional and require condition!



````python
# if-elif-else

age = 33

if age <= 9:
  print('If: 10살 이하')
elif age <= 19:
  print('Elif: 10대')
elif age <= 29:
  print('Elif: 20대')
elif age <= 39:
  print('Elif: 30대')
elif age <= 49:
  print('Elif: 40대')
elif age <= 59:
  print('Elif: 50대')
else:
  print('Else: 60대 이상')
````

**Nested If Statements**: Nested if can be useful when we need to expand several conditional statements. 



````python
# nested if

# case1:
x = 3
y = "Hello"

# case2:
# x = 3
# y = 'Wow'

# case3:
# x = 7
# y = 'Wow'

# case4:
# x = 3
# y = "Hello"

if x < 5:
  if y == "Hello":
    x = 0
  else:
    y = "Hi"
else:
  x = 10

print(f'x is {x} and y is {y}')
````

### Pass statement and Conditional expressions

A pass statement is a **null** statement - when it is executed, nothing happens. When a statement is syntatically required but you have nothing specific to do - then we need to use a 'pass' statement.
- `if`문에는 무엇이든 내용이 있어야 함: 만일 아무것도 실행할 내용이 없다면 `pass`를 사용해 자리 채우기를 할 수 있음


A **conditional expression** is a concise version of `if-else` statement: write if in **one line**


````python
i = 1

if i > 7:
  # pass
else:
  print(i)
  print('i is less than 7 - you are in "else" statement')
````

````python
# if-else statement 
a = 1
if a % 2 == 0:
  result = 'even'
else:
  result = 'odd'

print(result)
````

````python
# conditional expression
a = 1
result = 'even' if a % 2 == 0 else 'odd'
print(result)
````

## Loop statements (반복문)

**Loop** is a very powerful concept to execute a task in iterations:



### For Loop
**For loop** is used for repeating a task for a desired number of times.

```
for iterator in iterable:
  <statement>
```
- iterator: a variable that is used to iterate through the collection 
- iterable: a sequence or collection through which iteation happens in for loops 
  - list ``['a','b','c']``, range ``range(3)``, tuple ``(1,2,3)``, ... 
- statement: code block to be executed during each iteration in the loop - must **indented**
- `iterable`에 해당하는 항목에 대해 코드를 반복해서 실행함: 정해진 횟수만큼 반복을 수행
- `iterator in iterable`: `iterable`에 속한 항목들을 차례대로 `iterator`라는 이름의 변수로 받아옴을 의미

````python
# list, range()
a = [0,1,2]
print(a)

b = range(3)
print(b)
print(tuple(b))

print(b[0])
print(b[1])
print(b[2])

c = range(1, 11, 2)
print(c)
print(list(c))
````

````python
# 1: without for loop, print each element of ['a','b','c']

lst = ['a','b','c']

val = lst[0]
print(val)

val = lst[1]
print(val)

val = lst[2]
print(val)

# 1: with for loop, print each element of ['a','b','c']

lst = ['a','b','c']
for val in lst:
  print(val)
````

````python
# 2

for iter in ['Alice','Bob','Charlie']:
  print(f'Hi! {iter}')

print('=====')
for iter in range(5):
  i_str = 'Hi'*iter
  print(i_str)
````

````python
# 3: compute a sum and product of all numbers from 1 to 4

sum = 0 
for i in range(1,5):
  sum = sum+i
  print(sum)
print(f'Sum of all numbers from 1 to 4 is {sum}')

print('=====')

product = 1
for i in range(1,5):
  product *= i
  print(product)
print(f'Product of all numbers from 1 to 4 is {product}')
````

````python
# 4: given a sentence, print each word

my_str = 'One year has 12 months, 52 weeks and 365 days'
my_lst = my_str.split()
print(my_lst)

for i in my_lst:
  print(i)
````

````python
# 5: given a dictionary, print each key and value 

my_dict = {'name': 'Peter', 'birth': ['April', 9], 'age': 27, 'year': 1995, 'Course': 'Python'}
print(my_dict.keys())

print('=====')
for key in my_dict.keys():
  print(f'Key: {key} and Value: {my_dict[key]}')
````

* 반복문을 수행하는 항목이 여러 개의 값을 가진다면 여러 개의 값을 받아올 수 있음:

````python
# 5: given a dictionary, print each key and value 

my_dict = {'name': 'Peter', 'birth': ['April', 9], 'age': 27, 'year': 1995, 'Course': 'Python'}

print('=====')

for (key, value) in my_dict.items():
  print(f'Key: {key} and Value: {value}')
````

* for 문을 중첩하여 사용하면 계층적인 반복을 수행할 수 있음

````python
# Nested for loop: 1부터 5까지 구구단을 출력

num = 5
for i in range(1,num+1):
  print(f'Line {i}:', end=' ')
  
  for j in range(1,10):
    print(i*j, end=' ')

  print('')
````

### While Loop

**While Loop** is another powerful repetition structure in Python. 

```
while <condition>:
  <statement>
```

It executes the ``<statement>`` as long as the ``<condition>`` is satisfied.
- 특정 조건 `<condition>`을 만족하는 동안 코드를 반복해서 실행
- 반복 횟수가 명확히 정해져 있지 않을 때 사용

````python
# 1 
i = 1
while i < 6:
  print(i)
  i += 1
````

````python
# 2 
i = 10
while i > 0:
  if i % 2 == 0:
    print(f'{i} is an even number')

  i -= 1
````

````python
# 3: calculate sum of numbers until the user enters zero
total_sum = 0

number = int(input('Enter an integer: '))

# add numbers until next number is zero
while number != 0:
  total_sum += number
  number = int(input('Enter a next integer: ')) # need to take next integer!

# after while loop: print the total sum
print(f'Total sum is {total_sum}')
````

### Break, Continue Statements

* A **break** statement terminates the loop (전체 반복문을 탈출)
* A **continue** statement skips a code block in the loop for the current iteration (현재 실행 중인 반복만 끝내고 다음 반복으로 이동)
- 아래의 코드가 어떤 동작을 수행할 지 예상해 보고 실행하여 결과를 확인해보세요 

````python
i = 10

while i > 0:
  i = i-1
  if i > 7:
    continue
  elif i < 3:
    break
  else:
    print(i)
````

````python
for i in range(20):
  if i % 3 != 0:
    continue
  if i == 15:
    break

  print(i)
````

````python
count_even = 0
count_odd = 0

while True:
  val = int(input('Enter an integer: '))
  if val == 10:
    break
  elif val % 2 == 0: 
    count_even += 1
  else: 
    count_odd += 1

print(f'Finish the loop: number of evens {count_even} and number of odds {count_odd}')
````
